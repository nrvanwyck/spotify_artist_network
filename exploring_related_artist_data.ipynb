```python
import json

with open('artist_dict.json', 'r') as file:
    artist_dict = json.load(file)
```

```python
adjacency_dict = {}
for artist_uri in artist_dict:
    if 'related_artists' in artist_dict[artist_uri]:
        for related_artist_uri in artist_dict[artist_uri]['related_artists']:
            if artist_uri not in adjacency_dict:
                adjacency_dict[artist_uri] = set()
            adjacency_dict[artist_uri].add(related_artist_uri)
            if related_artist_uri not in adjacency_dict:
                adjacency_dict[related_artist_uri] = set()
            adjacency_dict[related_artist_uri].add(artist_uri)
```

```python
genre_dict = {}
for artist_uri in adjacency_dict:
    genre_counter = {}
    for genre in artist_dict[artist_uri]['genres']:
        genre_counter[genre] = 0
        for adjacent_artist_uri in adjacency_dict[artist_uri]:
            if genre in artist_dict[adjacent_artist_uri]['genres']:
                genre_counter[genre] += 1
    genre_dict[artist_uri] = {genre for genre in genre_counter if 
                              genre_counter[genre] == 
                              max(genre_counter.values())}
    
previous_counter = None
while True:
    counter = 0
    for artist_uri in genre_dict:
        if len(genre_dict[artist_uri]) > 1:
            counter += 1
            genre_counter = {genre: 0 for genre in genre_dict[artist_uri]}
            for adjacent_artist_uri in adjacency_dict:
                if len(genre_dict[adjacent_artist_uri]) == 1:
                    genre = list(genre_dict[adjacent_artist_uri])[0]
                    if genre in genre_counter:
                        genre_counter[genre] += 1
            genre_dict[artist_uri] = {genre for genre in genre_counter if 
                                      genre_counter[genre] == 
                                      max(genre_counter.values())}
    if counter == previous_counter:
        break
    else:
        previous_counter = counter
        
for artist_uri in artist_dict:
    print(artist_dict[artist_uri]['name'])
    print(genre_dict[artist_uri])
    print()
```
